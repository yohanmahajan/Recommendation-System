```python
import pandas as pd
import numpy as np
import time
```

```python
df = pd.read_csv('ratings.csv')
```

```python
df
```

```python
movies = pd.read_csv('movies.csv')
```

```python
movies
```

```python
df = pd.merge(df,movies,on='movieId')
```

```python
df
```

```python
# Getting the average rating for every movie
avg_rating_df = pd.DataFrame(df.groupby('title')['rating'].mean())
```

```python
avg_rating_df
```

```python
# calculating the total number of people watched every movie (assuming that every person watched the movie, rated it)
avg_rating_df['no_of_ratings'] = df.groupby('title')['rating'].count()
```

```python
avg_rating_df.head()
```

```python
# creating a table with rows as the viewer and columns as the movie, where every cell corresponds to the rating given 
# by the respective user to the respective movie. 
um_rating = df.pivot_table(index='userId',columns='title',values='rating')
```

```python
um_rating
```

```python
rec_mov = pd.DataFrame()
```

```python
def recommend(userID):
    global rec_mov
    user_rating = []
    for i in um_rating.columns:  # picking up movies rated well by the user
        if um_rating[i][userID] >= 4:
            user_rating.extend([i])
            
    user_rating_final = []
    for movie in user_rating:  # picking movies with >100 ratings
        if avg_rating_df.loc[movie]['no_of_ratings'] > 100:
            user_rating_final.extend([movie])
    
    #rec = pd.DataFrame()
    rec = []
    for i in user_rating_final:
        rating = um_rating[i]
        similar = um_rating.corrwith(rating)
        corr = pd.DataFrame(similar,columns=['correlation'])
        corr.dropna(inplace=True)
        for j in corr.index:
            if corr.loc[j]['correlation'] >= 0.95 and j not in rec and j not in user_rating: # assuming that 2 users with correlation higher than 0.95 are very similar
                rec.extend([j])
                r = {'movie': [j], 'correlation': [round(corr.loc[j]['correlation'],2)]}
                record = pd.DataFrame(data=r)
                rec_mov = rec_mov.append(record, ignore_index = True)
                rec_mov = rec_mov.sort_values(by=['correlation'], ascending=False)
    return user_rating, rec_mov['movie'] #returning a dataframe of recommended movies with correlation
```

```python
# start_time = time.time()
liked, rec = recommend(2)
# print("--- %s seconds ---" % (time.time() - start_time))
```
