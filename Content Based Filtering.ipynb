```python
movies = movies.drop_duplicates(subset='title', keep="first")
```

```python
movies.reset_index(inplace = True)
```

```python
df = pd.merge(df,movies,on='movieId')
```

```python
df
```

```python
# Getting the average rating for every movie
avg_rating_df = pd.DataFrame(df.groupby('title')['rating'].mean())
```

```python
avg_rating_df
```

```python
# calculating the total number of people watched every movie (assuming that every person watched the movie, rated it)
avg_rating_df['no_of_ratings'] = df.groupby('title')['rating'].count()
```

```python
avg_rating_df.head()
```

```python
# creating a table with rows as the viewer and columns as the movie, where every cell corresponds to the rating given 
# by the respective user to the respective movie. 
um_rating = df.pivot_table(index='userId',columns='title',values='rating')
```

```python
um_rating
```

```python
for i in range(len(movies)):
    movies['genres'][i] = movies['genres'][i].split('|')
```

```python
movies
```

```python
movies.set_index(['title'], inplace=True)
```

```python
movies
```

```python
def recommend(userID):
    user_rating = []
    for i in um_rating.columns:  # picking up movies rated well by particular user
        if um_rating[i][userID] >= 4:
            user_rating.extend([i])
            
    liked_genre = []
    
    for i in user_rating:
        if movies.loc[i]['genres'] not in liked_genre:
            liked_genre.append(movies.loc[i]['genres'])
            
    rec = []
    
    for i in movies.index:
        if movies.loc[i]['genres'] in liked_genre:
            if i not in user_rating:
                rec.append(i)
                
    return user_rating, rec
```

```python
liked, rec = recommend(2)
```
