```python
movies = pd.read_csv('movies.csv')
```

```python
df = pd.merge(df,movies,on='movieId')
```

```python
# Getting the average rating for every movie
avg_rating_df = pd.DataFrame(df.groupby('title')['rating'].mean())
```

```python
avg_rating_df
```

```python
# calculating the total number of people watched every movie (assuming that every person watched the movie, rated it)
avg_rating_df['no_of_ratings'] = df.groupby('title')['rating'].count()
```

```python
avg_rating_df.head()
```

```python
# creating a table with rows as the viewer and columns as the movie, where every cell corresponds to the rating given 
# by the respective user to the respective movie.
um_rating = df.pivot_table(index='title',columns='userId',values='rating')
```

```python
um_rating
```

```python
rec_mov = pd.DataFrame()
```

```python
def recommend(userID):
    global rec_mov
    
    # picking users with similar taste in movies like that of the prime user under consideration
    rating = um_rating[userID]
    similar = um_rating.corrwith(rating) 
    corr = pd.DataFrame(similar,columns=['correlation'])
    
    corr.dropna(inplace=True)
    corr.drop([2], axis = 0, inplace=True)
    
    for j in corr.index:
        if corr.loc[j]['correlation'] < 0.95: # assuming that 2 users with correlation higher than 0.95 are very similar
            corr.drop((j), axis = 0, inplace=True)
            
    user_rating = []
    for i in corr.index:  # picking up movies rated well by particular user
        for j in um_rating.index:
            if um_rating[i][j] >= 4:
                user_rating.extend([j])
    
    rec = []
    for i in user_rating:
        if pd.isna(um_rating[userID][i]):
            rec.append(i)
            
    liked = []
    for i in um_rating.index:  # picking up movies rated well by the user
        if um_rating[userID][i] >= 4:
            liked.extend([i])
            
    return liked, rec
```

```python
liked, rec = recommend(2)
```
